```python
import pandas as pd
##read csv
autos = pd.read_csv('autos.csv')
autos.head()
```

```python
##defining the input and target variables
x = autos[['cylinders', 'displacement','horsepower','weight','acceleration','year']]
y = autos['mpg']

from sklearn.model_selection import train_test_split

##splitting the data
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = .2)
```

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np
## defining # of trees
ntree = [100,300,500,800,1000]
##extracting tree
n = len(ntree)
##definging the dataframe to store results
results = pd.DataFrame({'Numoftrees': np.repeat(np.nan, n ), 'RMSE': np.repeat(np.nan, n), 'MAE': np.repeat(np.nan,n)})

for i in range(0,n):
    ##building rf model
    rf = RandomForestRegressor(n_estimators = ntree[i]).fit(xtrain,ytrain)
    ##predict on rest
    rfpred = rf.predict(xtest)
    ##store results
    results['Numoftrees'][i] = ntree[i]
    results['RMSE'][i] = np.sqrt(np.mean((rfpred - ytest)**2))
    results['MAE'][i] = np.mean(abs(rfpred - ytest))
    
results
```
