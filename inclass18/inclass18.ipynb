```python
import pandas as pd
##read csv
churn = pd.read_csv('Customer_Churn.csv')
churn.head()
```

```python
##Question 1) A, true
##Question 2) B, false
```

```python
import numpy as np
##changing labels ot 0 and 1
churn['churn_numb'] = np.where(churn['Churn']=='No', 0 , 1)
churn.head()
```

```python
##defining the input and target variables
x = churn[['SeniorCitizen','tenure','MonthlyCharges']]
y = churn['churn_numb']

from sklearn.model_selection import train_test_split

##spliting the data
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = .2, stratify = y)
```

```python
##building the logistic regression model
from sklearn.linear_model import LogisticRegression
logit_md = LogisticRegression().fit(xtrain, ytrain)
##predicting the test dataset
logit_pred = logit_md.predict_proba(xtest)[:,1]
logit_pred_label = np.where(logit_pred < .35, 0, 1)

from sklearn.metrics import confusion_matrix, accuracy_score, recall_score
print('the confusion matrix is', confusion_matrix(ytest, logit_pred_label))
print('the accuracy is', accuracy_score(ytest, logit_pred_label))
print('the recall is', recall_score(ytest, logit_pred_label))
```

```python
##building random forest
from sklearn.ensemble import RandomForestClassifier

rf_md = RandomForestClassifier().fit(xtrain,ytrain)

##predicting the test dataset
rf_pred = rf_md.predict_proba(xtest)[:,1]
rf_pred_label = np.where(rf_pred < .35, 0, 1)

print('the confusion matrix is', confusion_matrix(ytest, rf_pred_label))
print('the accuracy is', accuracy_score(ytest, rf_pred_label))
print('the recall is', recall_score(ytest, rf_pred_label))
```

```python
##gradient boosting model 
from sklearn.ensemble import GradientBoostingClassifier

gbc_md = GradientBoostingClassifier().fit(xtrain, ytrain)

##predicting the test dataset
gbc_pred = rf_md.predict_proba(xtest)[:,1]
gbc_pred_label = np.where(gbc_pred < .35, 0, 1)

print('the confusion matrix is', confusion_matrix(ytest, gbc_pred_label))
print('the accuracy is', accuracy_score(ytest, gbc_pred_label))
print('the recall is', recall_score(ytest, gbc_pred_label))
```
