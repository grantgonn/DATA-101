```python
import pandas as pd
##reading csv file
autos = pd.read_csv('Automobile_data.csv')
autos.head()
```

```python
import matplotlib.pyplot as plt
## creaing scatterplot
plt.plot('horsepower', 'price', 'o', data = autos)
plt.xlabel('Horsepower')
plt.ylabel('price')
plt.grid()
plt.show()
##the plot shows there seems to be a linear relationship between horsepower and price
```

```python
import statsmodels.formula.api as smf
##building the lm
lm_md = smf.ols(formula = 'price ~ horsepower', data = autos).fit()
##extracting model results
lm_md.summary()
##For one unit of increase horsepower the price increases by 208.67 on average 
```
